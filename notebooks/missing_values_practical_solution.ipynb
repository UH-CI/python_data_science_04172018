- What are the numbers of the rows for which the value of the column `specialty` is "NURSE PRACTITIONER" and the value of spending is lower than $5000
  - How can you remove those lines?


- Delete rows that have more than `4` values

- Replace the missing values for the columns `nb_beneficiaries` and `spending`  with their respective medians and the missing values in `specialty` with the the most frequent specialty




```python
import pandas as pd
spending_practical_df = pd.read_table("data/missing_values_practical.tsv", index_col="unique_id", dtype={'doctor_id': "object"})
```

```python
spending_practical_df.isnull().sum()
```

```python
spending_practical_df["specialty"].value_counts()
```

```python
spending_practical_df[(spending_practical_df["specialty"] == "NURSE PRACTITIONER") & (spending_practical_df["spending"] < 5000)]
```

```python
spending_practical_df.dropna(axis = 0, thresh=4, inplace=True)
```

```python
data = {  "spending": spending_practical_df["spending"].median(), 
          "nb_beneficiaries": spending_practical_df["nb_beneficiaries"].median(), 
         "specialty":   "UNKNOWN"}

spending_practical_df.fillna(data, inplace=True)

spending_practical_df
```
